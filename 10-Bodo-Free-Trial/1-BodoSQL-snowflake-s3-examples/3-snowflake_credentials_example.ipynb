This example shows how to use bodo platform utils and secrets to connect to snowflake.

```python
%autopx
```

# Loading Snowflake Credentials

```python
from bodo_platform_utils import secrets
import bodo
import pandas as pd
```

```python
sf_credentials = secrets.snowflake_credentials()
```

```python
if bodo.get_rank() == 0:
    display(sf_credentials)
```

# Reading from Snowflake

```python
def get_snowflake_url():
    url = f"snowflake://{sf_credentials['username']}:{sf_credentials['password']}@{sf_credentials['account']}/{sf_credentials['database']}/{sf_credentials['schema']}?warehouse={sf_credentials['warehouse']}"
    return url


@bodo.jit
def read_snowflake(table_name, conn_str):
    query = f"SELECT * FROM {table_name}"
    df = pd.read_sql(query, conn_str)
    return df


url = get_snowflake_url()
```

```python
table_name = "CUSTOMER"
df = read_snowflake(table_name, url)
df
```
